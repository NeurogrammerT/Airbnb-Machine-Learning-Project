```python
# Dependencies
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv("train_users_2.csv")
train_df
```

```python
train_df_2 = train_df.filter(["id", "date_account_created", "date_first_booking"], axis=1)
train_df_2.head()
```

```python
new_train_df = train_df_2.dropna()
new_train_df.head()
```

```python
new_train_df['date_account_created'] = pd.to_datetime(new_train_df['date_account_created'])
new_train_df['date_first_booking'] = pd.to_datetime(new_train_df['date_first_booking'])
```

```python
new_train_df["time_to_booking"] = (new_train_df["date_first_booking"] - new_train_df["date_account_created"]).dt.days
new_train_df.head()
```

```python
new_train_df.to_csv(r'C:\Users\zineb\train_df.csv', index = None, header=True)
```

```python
time_df = pd.read_csv("new_train_df.csv")
time_df.head()
```

```python
time_df.to_csv('output.csv', index=False, header=None)
```

```python
# Create a list to store the data
acct_book = []

# For each row in the time_difference column to bucket the booking time from account creation into buckets
for row in time_df['time_difference']:
    if int(row) > 180:
        acct_book.append('more than 6 months')
    elif int(row) > 90:
        acct_book.append('within 6 months')
    elif int(row) > 30:
        acct_book.append('within three months')
    elif int(row) > 7:
        acct_book.append('within a month')
    elif int(row) > 0:
        acct_book.append('within a week')
    elif int(row) == 0:
        acct_book.append('instantly')
    else:
        # Append a failing acct_book
        acct_book.append('Failed')
        
# Create a column from the list
time_df['booked_within'] = acct_book
time_df.head()
```

```python
booked_within = time_df.groupby(['booked_within'], as_index=False).agg('count')
booked_within_df = booked_within.drop(['date_account_created', 'date_first_booking', 'time_to_booking', 'time_difference'], axis=1)
booked_within_df.sort_values('id', ascending=False)
```

```python
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]
plot = plt.pie(
    booked_within_df['id'],
    labels = booked_within_df['booked_within'],
    colors = colors,
    shadow = True,
    startangle=90,
    autopct='%1.1f%%',
    explode=(0.10, 0, 0, 0, 0.10, 0),
)
plt.axis('equal')
plt.tight_layout()
plt.title('Booking timeframes statistics from account creation')
plt.savefig('pie.png')
```
