```python
# Dependencies and Setup

import pandas as pd
import numpy as np

# Files to Load

train_file = "train_file_427_v2.csv"
```

```python
# Read Train File and store into Pandas data frame
train_read = pd.read_csv(train_file)
train_data = train_read.rename(index=str, columns={"id": "user_id"})
train_data.head()
```

```python
# developed the booking = 0/1 based on booking date
train_data['stringdate'] = train_data.date_first_booking.astype(str)
train_data.loc[train_data['stringdate'] == 'nan', 'booked'] = 0
train_data.loc[train_data['stringdate'] != 'nan', 'booked'] = 1

print(train_data.booked.value_counts())
train_data.head()
```

```python
train_data.to_csv('train_data_booked.csv')
```
