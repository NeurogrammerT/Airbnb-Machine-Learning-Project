```python
# Dependencies
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
sessions_df = pd.read_csv("sessions.csv")
sessions_df.head()
```

```python
sessions_2 = sessions_df.rename(columns={"user_id": "id"})
sessions_2.head()
```

```python
sessions_2_df = sessions_2.dropna()
sessions_3_df = sessions_2_df.drop(["device_type"], axis=1)
sessions_3_df.head()
```

```python
train_df = pd.read_csv("train_file_427_v2_2.csv")
train_df.head()
```

```python
new_train_df = train_df[["id", "booking"]]
new_train_df.head()
```

```python
merged_df = pd.merge(sessions_3_df, new_train_df, on="id")
merged_df.head()
```

```python
sum_df = merged_df.groupby(['id', 'action', 'booking'], as_index = False)['secs_elapsed'].sum()
sum_df.head()
```

```python
non_booking = sum_df[sum_df['booking'] == 0].index
sum_df.drop(non_booking , inplace=True)
```

```python
sum_df.to_csv("sum_df.csv", index=True, header=True)
```

```python
sum_df_non = merged_df.groupby(['id', 'action', 'booking'], as_index = False)['secs_elapsed'].sum()
sum_df_non.head()
```

```python
booking = sum_df_non[sum_df_non['booking'] == 1].index
sum_df_non.drop(booking, inplace=True)
sum_df_non.head()
```

```python
sum_df_non.to_csv("sum_df_non.csv", index=True, header=True)
```

```python
conct_df = pd.concat([sum_df, sum_df_non])
conct_df.head()
```

```python
conct_df.loc[conct_df.booking == 1, "booking"] = "booked"
conct_df.loc[conct_df.booking == 0, "booking"] = "non_booked"
conct_df.head()
```

```python
conct_df.to_csv("conct_df.csv", index=True, header=True)
```
