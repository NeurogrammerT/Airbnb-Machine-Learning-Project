```python
import pandas as pd
import numpy as np
```

```python
sessions_df = pd.read_csv("airbnb-recruiting-new-user-bookings/sessions.csv")
sessions_df.head(10)
```

```python
total_sessions_users = sessions_df['user_id'].nunique()

print(f'Total Unique User Ids: {total_sessions_users}')  
```

```python
user_action_count = sessions_df.groupby("user_id").count()

user_action_count = pd.DataFrame(user_action_count)

user_action_count = user_action_count.sort_values(by="action", ascending=False)

user_action_count = user_action_count.drop(['device_type','secs_elapsed'],axis=1)

user_action_count.reset_index(inplace=True)

print(len(user_action_count))
user_action_count.head(20)
```

```python
total_user_time = sessions_df.groupby("user_id").sum()

total_user_time = pd.DataFrame(total_user_time)

total_user_time = total_user_time.sort_values(by="secs_elapsed", ascending=False)

total_user_time['hrs_elapsed'] = (total_user_time['secs_elapsed']/60)/60

total_user_time.reset_index(inplace=True)

print(len(total_user_time))
total_user_time.head(20)
```

```python
sessions_df2 = pd.DataFrame(sessions_df)

sessions_df2['Desktop']=0
sessions_df2['Tablet_LapTop']=0
sessions_df2['Phone']=0

sessions_df2['Desktop'][sessions_df2['device_type']=='Windows Desktop']=1
sessions_df2['Desktop'][sessions_df2['device_type']=='Linux Desktop']=1
sessions_df2['Desktop'][sessions_df2['device_type']=='Mac Desktop']=1

sessions_df2['Tablet_LapTop'][sessions_df2['device_type']=='iPad Tablet']=1
sessions_df2['Tablet_LapTop'][sessions_df2['device_type']=='Tablet']=1
sessions_df2['Tablet_LapTop'][sessions_df2['device_type']=='-unknown-']=1
sessions_df2['Tablet_LapTop'][sessions_df2['device_type']=='Chromebook']=1

sessions_df2['Phone'][sessions_df2['Desktop'] + sessions_df2['Tablet_LapTop'] == 0]=1
#sessions_df2['Phone'][sessions_df2['Tablet_LapTop']+[sessions_df2['Desktop'] == 0]=1


sessions_df2.tail(500)
```

```python
#total_user_time = total_user_time.sort_values(by="secs_elapsed", ascending=False)
#user_action_count = sessions_df.groupby("user_id").count()

sessions3 = sessions_df2.groupby('user_id').sum()

sessions3 = sessions3.drop(['secs_elapsed'], axis=1)

sessions3.reset_index(inplace=True)
sessions3.head(500)
```

```python
#train_sessions = pd.merge(train_df,sessions_aggr,how='inner',on='id')

sessions_agg = pd.merge(sessions3,total_user_time, how='inner', on='user_id')

sessions_agg = pd.merge(sessions_agg,user_action_count, how='inner', on='user_id')

sessions_agg = sessions_agg.drop(['secs_elapsed'], axis=1)

sessions_agg.head()
```

```python
sessions_agg.to_csv("sessions_agg_final.csv",index=False)
```
