```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv("train_data_final.csv")
train_data.head()
```

```python
from sklearn.model_selection import train_test_split

X = train_data.drop("country_destination", axis=1)
y = train_data["country_destination"]

print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y, train_size=.7)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=1000)
model
```

```python
model.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
from sklearn.multiclass import OneVsOneClassifier
from sklearn.svm import LinearSVC

model2A = OneVsOneClassifier(LinearSVC())
```

```python
model2A.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {model2A.score(X_train, y_train)}")
print(f"Testing Data Score: {model2A.score(X_test, y_test)}")
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC

model2 = OneVsRestClassifier(LinearSVC())
```

```python
model2.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {model2.score(X_train, y_train)}")
print(f"Testing Data Score: {model2.score(X_test, y_test)}")
```
