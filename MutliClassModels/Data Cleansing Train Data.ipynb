```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
train_df = pd.read_csv("airbnb-recruiting-new-user-bookings/train_users_2.csv")
print(len(train_df))
train_df.head(20)
```

```python
train_df['stringdate'] = train_df.date_first_booking.astype(str)
train_df.loc[train_df['stringdate'] == 'nan', 'booked'] = 0
train_df.loc[train_df['stringdate'] != 'nan', 'booked'] = 1

train_df.head()
```

```python
train_df['age'].mean()
```

```python
train_df['age_revised'] = train_df['age']
#sessions_df2['Tablet_LapTop'][sessions_df2['device_type']=='iPad Tablet']=1
train_df['age_revised'][train_df['age']<20] = 20
train_df['age_revised'][train_df['age']>60] = 30 

train_df.head(100)
```

```python
train_df.isnull().sum()
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
train_df['age_3']=imputer.fit_transform(train_df[['age_revised']])

train_df.isnull().sum()
```

```python
train_df.head()
```

```python
list(train_df)
```

```python
train_2 = train_df.drop(['date_account_created','timestamp_first_active',\
                         'date_first_booking',\
                         'age',\
                         'signup_method',\
                         'signup_flow',\
                         'language',\
                         'affiliate_channel',\
                         'affiliate_provider',\
                         'first_affiliate_tracked',\
                         'signup_app',\
                         'first_device_type',\
                         'first_browser',\
                         'stringdate',\
                          'age_revised'], axis=1)

train_2.head()
```

```python
train_2.gender.value_counts()
```

```python
gender_dummies = pd.get_dummies(train_2.gender)
gender_dummies.head()
```

```python
train_2 = pd.concat([train_2, gender_dummies], axis=1)

train_2 = train_2.drop(['gender'], axis=1)
train_2 = train_2.rename(columns={'id':'user_id'})

print(len(train_2))
train_2.head()
```

```python
session_agg = pd.read_csv("sessions_agg_final.csv")

session_agg.head()
```

```python
train_data = pd.merge(session_agg ,train_2, how='inner', on='user_id')
print(len(train_data))
train_data.head()
```

```python
train_data = train_data.drop(['user_id'], axis=1)
train_data.head()
```

```python
train_data.country_destination.value_counts()
train_data.isnull().sum()
```

```python
train_data.to_csv("train_data_final.csv",index=False)
```
