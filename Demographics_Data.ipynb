```python
# Dependencies and Setup

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Files to Load

demo_file = "age_gender_bkts.csv"
```

```python
# Read Age File and store into Pandas data frame
demo_data = pd.read_csv(demo_file)
demo_data.head()
```

```python
age_gender = demo_data.groupby(['age_bucket', 'gender'], as_index=False).sum()
age_gender.sort_values(by=['age_bucket'])
age_gender.drop(['year'], axis=1)
```

```python
color = ["#006D2C","#74C476"]
#ax = age_gender['age_bucket'].plot(kind='bar', stacked=True, color=color)
ax = age_gender[['age_bucket', 'gender','population_in_thousands']].plot(kind='bar', stacked=False,
              figsize=(15,7), color=color, fontsize=13);
```
