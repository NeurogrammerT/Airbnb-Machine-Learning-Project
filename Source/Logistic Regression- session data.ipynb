```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from sklearn.metrics import confusion_matrix
```

```python
#import Data
train = pd.read_csv("train_file_427.csv")

train.head()
train.info()

train1 = train.loc[train['session_flag_1'] == 1]
train1.session_flag_1.value_counts()
```

```python
mykeeplist = list(train1.columns)[1:]
mykeeplist
```

```python
train_data = train1[mykeeplist]
train_data.head()
```

```python
list(train_data.columns)
```

```python

# del train_data['pending_1']
del train_data['requested_u']
#del train_data['requested_1']
#del train_data['dashboard_1']
del train_data['dashboard_u']
#del train_data['edit_1']
del train_data['edit_u']

del train_data['set_user_u']
del train_data['verify_u']
del train_data['confirm_email_u']
del train_data['travel_plans_current_u']
del train_data['pending_u']
del train_data['identity_u']
del train_data['cancellation_policies_u']
del train_data['show_personalize_u']
del train_data['secs_elapsed_u']
del train_data['session_flag_u']
del train_data['Mac_PC']
```

```python
list(train_data.columns)
```

```python
#Designate X (Data) and y (Target)
# where YYYYY= to the column label that is the target

#drops the column with the target data from the data
X = train_data.drop("booked", axis=1) 

#assigns column with the target data from the data to y
y = train_data["booked"]

#verify that target column has been dropped
print(X.shape, y.shape)
```

```python
#spllit data into training and testing data

from sklearn.model_selection import train_test_split

#Adjust Trainting size >> Must be a number between 0-1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y, train_size=.75)
```

```python
#create logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
#Fit Model
classifier.fit(X_train, y_train)
```

```python
# Print Model Scores

print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score:  {classifier.score(X_test, y_test)}")
```

```python
# Predicitons
#### Ensure Test data has session values
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")

pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```

```python
confusion_matrix(y_test, predictions)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
print(f"true did not book: {tn:3d}   fp: booked but didnt {fp:3d}")
print(f"fals did not book: {fn:3d}   tp: booked and did   {tp:3d}")
```
